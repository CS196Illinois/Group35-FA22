```python
from ast import Attribute
import re

myString = "<li>March 11 System Massmail: <a href=\”https://massmail.illinois.edu/massmail/28768.html\”>COVID-19 Policies</a></li><li>March 5 Massmail: <a href=\”https://massmail.illinois.edu/massmail/28774.html\”>An update on COVID-19</a></li><li>March 2 Massmail: <a href=\”https://massmail.illinois.edu/massmail/28770.html\”>COVID-19 Update: International Travel</a></li><li>January 31 Massmail: <a href=\”https://massmail.illinois.edu/massmail/28750.html\”>Update on coronavirus, January 31, 2020</a></li><li>January 29 Massmail: <a href=\”https://massmail.illinois.edu/massmail/28749.html\”>Update on coronavirus for the campus community</a></li><li>January 27 Massmail: <a href=\”https://massmail.illinois.edu/massmail/28748.html\”>Updated Coronavirus Information</a></li><li>January 23 Massmail: <a href=\”https://massmail.illinois.edu/massmail/28745.html\”>Coronavirus Precautions</a></li></ul>"

days = myString.split("<li>")

links = []

titles = []

for day in days:
    try:
      link = re.search("(?P<url>https?://[^\s]+)", day).group("url")
      links.append(link)
      titles.append(day.split(":")[0])
    except AttributeError:
      print("")
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

i = 0

for link in links:
  url = (link.split(".html")[0] + ".html")
  #url = "https://massmail.illinois.edu/massmail/1685707347.html"
  html = urlopen(url).read()
  soup = BeautifulSoup(html, features="html.parser")

  # remove all script and style elements
  for script in soup(["script", "style"]):
      script.extract()

  # get text
  text = soup.get_text()

  # break into lines and remove leading and trailing spaces
  lines = (line.strip() for line in text.splitlines())
  # break multi-headlines into a line each
  chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
  # remove blank lines
  text = '\n'.join(chunk for chunk in chunks if chunk)
  
  text = text.split("Powered by Webtools")[0]
  
  print(text)
  
  f = open("C://Users//nehav//Downloads//" + titles[i].split(" ")[0] + " " + titles[i].split(" ")[1] + ".txt", "w+", encoding='utf-8')
  f.write(text)
  f.close()

  i+=1
```
