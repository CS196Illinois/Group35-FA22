```python
import string
import os
import pandas as pd

def list_from_file(filename):
    myfile = open(filename, 'r')
    data = myfile.read().split()
    col = []
    for word in data:
        col.append(word)
    return col

def myhist(col):
    hist = {}
    for word in col:
        word = word.lower()
        word = word.strip(string.punctuation + string.whitespace)
        hist[word] = hist.get(word, 0)+1
    return hist

df = pd.DataFrame()

for filename in os.listdir('/Users/ayana/Downloads/cs124h/ayanap2/Project/Data/2020'):
    col = list_from_file("/Users/ayana/Downloads/cs124h/ayanap2/Project/Data/2020/" + filename)
    a = pd.DataFrame(myhist(col), index=[0])
    df = pd.concat([df, a])

for filename in os.listdir('/Users/ayana/Downloads/cs124h/ayanap2/Project/Data/2021'):
    col = list_from_file("/Users/ayana/Downloads/cs124h/ayanap2/Project/Data/2021/" + filename)
    a = pd.DataFrame(myhist(col), index=[0])
    df = pd.concat([df, a])

for filename in os.listdir('/Users/ayana/Downloads/cs124h/ayanap2/Project/Data/2022'):
    col = list_from_file("/Users/ayana/Downloads/cs124h/ayanap2/Project/Data/2022/" + filename)
    a = pd.DataFrame(myhist(col), index=[0])
    df = pd.concat([df, a])

df.dropna()
x = df.sum().to_dict()
y = pd.DataFrame(x, index = [0])
y.sort_values(by = [0], axis=1, ascending=False)
```
